## Predictive modeling of customer bookings

This Jupyter notebook includes some code to get you started with this predictive modeling task. We will use various packages for data manipulation, feature engineering and machine learning.

First, we must explore the data in order to better understand what we have and the statistical properties of the dataset.

## Step 1

### Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import RFE
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from imblearn.over_sampling import SMOTE
from category_encoders import TargetEncoder
import joblib
import os
```

### Load dataset

```python
df = pd.read_csv("../data/raw/customer_booking.csv", encoding="ISO-8859-1")
df.head()
```

## Step 2

### Initial Exploration

```python
print(df.info())
```

```python
# Remove all duplicate rows (keep the first occurrence)
df = df.drop_duplicates()
```

### Variables Distribution and Relationships Exploration

```python
# 2. Bar Charts for Categorical Features (8 plots)
plt.figure(figsize=(15, 12))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3, 3, i)
    sns.countplot(data=df, x=col)
    plt.xticks(rotation=45 if col in ['route', 'booking_origin'] else 0)
    plt.title(f'Count of {col}')
    if col in ['route', 'booking_origin']:
        plt.xticks([])  # Hide labels for high-cardinality columns
plt.tight_layout()
output_path = os.path.join(output_dir, "categorical_distributions.png")
plt.savefig(output_path, dpi=400, bbox_inches='tight')
plt.show()
```

```python
# Check class distribution in 'Revenue'
df_imbalanced = df['booking_complete'].value_counts(normalize=True) * 100

# Print class imbalance percentage
print("Class Imbalance in Bookings:")
print(df_imbalanced)
```

## Step 3 (Data Wrangling)

### Handle Outliers with Log Transformation

```python
for col in ['purchase_lead', 'length_of_stay']:
    df[col] = np.log1p(df[col])
```

## Step 4 (Feature Engineering)

```python
df['booking_weekday'] = df['flight_day'].map({'Mon': 1, 'Tue': 2, 'Wed': 3, 'Thu': 4, 'Fri': 5, 'Sat': 6, 'Sun': 7})
df['total_addons'] = df['wants_extra_baggage'] + df['wants_preferred_seat'] + df['wants_in_flight_meals']
df['is_peak_hour'] = df['flight_hour'].between(8, 20).astype(int)
df['lead_time_bucket'] = pd.qcut(df['purchase_lead'], q=5, labels=False)
df['stay_duration_cat'] = pd.qcut(df['length_of_stay'], q=3, labels=False)
```

### Target Encoding for High-Cardinality Categorical Variables ###

```python
# Step 6: Encode High-Cardinality Categorical Variables Using Target Encoding
target_encoder = TargetEncoder()
df[['route_encoded', 'booking_origin_encoded']] = target_encoder.fit_transform(df[['route', 'booking_origin']], df['booking_complete'])
df.drop(columns=['route', 'booking_origin'], inplace=True)
```

### Variable Manipulations ###

```python
# Step 7: Encode Remaining Categorical Variables
categorical_features = ['sales_channel', 'trip_type', 'flight_day']
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

## Step 5 (Modeling)

### Split Features and Target

```python
X = df.drop(columns=['booking_complete'])
y = df['booking_complete']
```

### Handle Class Imbalance with SMOTE

```python
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)
```

### Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced)
```

### Feature Selection using RFE

```python
selector = RFE(RandomForestClassifier(n_estimators=100, random_state=42), n_features_to_select=15)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)
```

### Get selected feature names

```python
selected_features = X.columns[selector.support_]
```

### Convert back to DataFrame

```python
X_train_selected = pd.DataFrame(X_train_selected, columns=selected_features)
X_test_selected = pd.DataFrame(X_test_selected, columns=selected_features)
```

### Build Preprocessing Pipeline

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), selected_features)
])
```

### Define Full Pipeline

```python
model_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'))
])
```

### Train Model

```python
model_pipeline.fit(X_train_selected, y_train)
```

### Evaluate Model

```python
y_pred = model_pipeline.predict(X_test_selected)
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred))
```

### Confusion Matrix

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

### Feature Importance

```python
importances = model_pipeline.named_steps['classifier'].feature_importances_
feature_importance_df = pd.DataFrame({'Feature': selected_features, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df.head(10))
```

```python
# Sort features by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance for Predicting Completed Bookings')
plt.gca().invert_yaxis()  # Invert y-axis to show the most important feature at the top
plt.savefig('../reports/figures/feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()
```
